```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

raw_data_count = pd.read_csv('recent-grads.csv')
recent_grads = raw_data_count.dropna()
recent_grads.head()


```

```python
recent_grads.plot(x ='Sample_size', y='Median', kind = 'scatter', title = 'Sample size vs Median')


```

```python
recent_grads.plot(x ='Sample_size', y='Unemployment_rate', kind = 'scatter', title = 'Sample size vs Unemployment rate')
```

```python
recent_grads.plot(x ='Full_time', y='Median', kind = 'scatter', title = 'Full time vs Median')
```

```python
recent_grads.plot(x ='ShareWomen', y='Unemployment_rate', kind = 'scatter', title = 'ShareWomen and Unemployment_rate')
```
