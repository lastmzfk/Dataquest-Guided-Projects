```python
autos.rename({'odometer': 'odometer_km'}, axis = 1, inplace = True)
```

```python
autos['odometer_km'] = (autos['odometer_km']
                       .str.replace('km', '')
                       .str.replace(',','')
                       .astype(int))
autos['price'] = (autos['price']
                 .str.replace('$', '')
                 .str.replace(',','')
                 .astype(int))

autos.head()
```

One thing that stands out from the exploration we did in the last screen is that the registration_year column contains some odd values:

The minimum value is 1000, before cars were invented
The maximum value is 9999, many years into the future

```python
autos = autos[autos['registration_year'].between(1910, 2019)]
autos['registration_year'].value_counts(normalize = True)
```

```python
print(autos['brand'].unique())
print(autos['brand'].describe())
brand_bool = autos['brand'].value_counts(normalize = True)
brands_explore = brand_bool[brand_bool > 0.05].index
print(brands_explore)

mean_price = {}
mean_mileage = {}

for brand in brands_explore:
    bo = autos[autos['brand'] == brand]
    brand_price = bo['price'].mean()
    brand_mileage = bo['odometer_km'].mean()
    mean_price[brand] = int(brand_price)
    mean_mileage[brand] = int(brand_mileage)

bmp_series = pd.Series(mean_price)
brand_mean = pd.DataFrame(bmp_series, columns = ['mean_price'])

bmm_series = pd.Series(mean_mileage)
brand_mean['mean_mileage'] = bmm_series

brand_mean
```

we aggregated across brands to understand mean price. We observed that in the top 6 brands, there's a distinct price gap.

Audi, BMW and Mercedes Benz are more expensive
Ford and Opel are less expensive
Volkswagen is in between
For the top 6 brands, let's use aggregation to understand the average mileage for those cars and if there's any visible link with mean price. While our natural instinct may be to display both aggregated series objects and visually compare them, this has a few limitations:

The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
